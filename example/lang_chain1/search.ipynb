```python
from langchain.document_loaders import YoutubeLoader #유튜브 로더
from langchain.embeddings import OpenAIEmbeddings #임베딩용
from langchain.chains import RetrievalQA #질문 답변 툴
from langchain.vectorstores import FAISS #**벡터 스토어**
from langchain_openai import ChatOpenAI
from pprint import pprint

import os, openai
from google.colab import userdata

os.environ["OPENAI_API_KEY"] = userdata.get("op")
openai.api_key=os.getenv("OPENAI_API_KEY") #결제 안해서 못 씀 코드만 작성 -> 결국 결제함

embeddings = OpenAIEmbeddings()

loader=YoutubeLoader.from_youtube_url('https://www.youtube.com/watch?v=HsonXuJs8-s') #주소 인자 넣기
#https://www.youtube.com/watch?v=HsonXuJs8-s 강의에서 사용한 주소
#찬물 샤워에 대한 내용임
documents=loader.load()

db=FAISS.from_documents(documents,embeddings)
retriever=db.as_retriever()

llm=ChatOpenAI(model_name='gpt-4')
qa=RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriever,return_source_documents=True)

query="What are the benefits of cold showers?"
#유튜브 내용을 기반으로 찬물 샤워에 대한 장점을 찾아줌
result =qa({"query": query})
pprint(result['result'])
```

```python
query="What are the benefits of cold showers? 한국말로 답해줘"
result =qa({"query": query})
pprint(result['result'])
```

```python
loader=YoutubeLoader.from_youtube_url('https://www.youtube.com/watch?v=4-cmJPA2Klg',language='ko') #주소 인자 넣기, 한국 뉴스니까 한국 언어
#https://www.youtube.com/watch?v=4-cmJPA2Klg 강의에서 사용한 주소
#"먹고 살기 힘들어"…대한민국 '허리'가 위험하다 / YTN
documents=loader.load()

db=FAISS.from_documents(documents,embeddings)
retriever=db.as_retriever()

llm=ChatOpenAI(model_name='gpt-4')
qa=RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriever,return_source_documents=True)

query="대강 무슨 내용이야?"
#유튜브 내용을 기반으로 내용 요약
result =qa({"query": query})
pprint(result['result'])
```

```python
#계속 같은거 하니까 함수 만들기
def ask_youtube(url,q,language='en'):
  loader=YoutubeLoader.from_youtube_url(url,language=language)
  documents=loader.load()

  db=FAISS.from_documents(documents,embeddings)
  retriever=db.as_retriever()

  llm=ChatOpenAI(model_name='gpt-4')
  qa=RetrievalQA.from_chain_type(llm=llm,
                                 chain_type="stuff",
                                 retriever=retriever,
                                 return_source_documents=True)

  result =qa({"query": q})
  return result
```
