```python
import os
from google.colab import userdata
os.environ["HUGGINGFACEHUB_API_TOKEN"] = userdata.get('hug')
```

```python
import langchain
from langchain_community.llms import HuggingFaceHub
repo_id="google/flan-t5-xxl"

llm = HuggingFaceHub(repo_id=repo_id, huggingfacehub_api_token=userdata.get('hug'),model_kwargs={"temperature":0.5, "max_length":64})
```

```python
from langchain_community.llms import HuggingFacePipeline
hf = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    pipeline_kwargs={"max_new_tokens": 10},
)
'''
위에 hub로 하려고 했는데 밑에 call하는 부분에서 아예 작동을 안해버림
서버랑 연결 어쩌구 하는 문제 같은데 해결할 방법을 모르겠음
토큰을 read로 해야되나? 내 hub에 있는 모델만 가능한건? 이유를 알고싶음
'''
```

```python
print(1)
print(hf("Tell me a joke"))
print(2)
#주의 사항 invoke를 사용하면 무제한 call을 하여 제한 토큰을 순식간에 사용해버림
#그냥 invoke 없이 사용하면 문제 없는 듯?
```
