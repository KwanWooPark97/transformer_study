```python
from transformers import AutoTokenizer
from bertviz.transformers_neuron_view import BertModel
from bertviz.neuron_view import show

model_ckpt = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = BertModel.from_pretrained(model_ckpt)
text = "time flies like an arrow"
show(model, "bert", tokenizer, text, display_mode="light", layer=0, head=8)
```

```python
inputs = tokenizer(text, return_tensors="pt", add_special_tokens=False)
inputs.input_ids
'''
"time flies like an arrow"에서 단어별로 임베딩 된 값을 의미
'''
```

```python
from torch import nn
from transformers import AutoConfig

config = AutoConfig.from_pretrained(model_ckpt)
token_emb = nn.Embedding(config.vocab_size, config.hidden_size)
token_emb
'''
30522개의 단어가 임베딩 되어있음 임베딩 벡터는 768차원 보유
'''
```

```python
inputs_embeds = token_emb(inputs.input_ids)
inputs_embeds.size()

'''
BATCH,SEQ,EMD
5개 데이터 보면 위에서 본 벡터랑 같을듯
'''
```

```python
import torch
from math import sqrt

query = key = value = inputs_embeds
dim_k = key.size(-1)
scores = torch.bmm(query, key.transpose(1,2)) / sqrt(dim_k)#정규화 하려고 제곱근으로 나눔
scores.size()

'''
bmm은 행렬 곱의미
batch seq, hidden 차원을 가지고있음
q*k 이고 차원이 [seq,hidden] 이므로 k를 전치함 [hidden,seq]로 그리고 bmm은 batch 순서대로 곱해줌 따라서 앞에 batch 차원은 빼고 생각함

'''
```

```python
import torch.nn.functional as F

weights = F.softmax(scores, dim=-1)
weights.sum(dim=-1)
```

```python
attn_outputs = torch.bmm(weights, value)
attn_outputs.shape
```

```python
def scaled_dot_product_attention(query, key, value):
    dim_k = query.size(-1)
    scores = torch.bmm(query, key.transpose(1, 2)) / sqrt(dim_k)
    weights = F.softmax(scores, dim=-1)
    return torch.bmm(weights, value)
'''
위에서 한 코드들을 하나의 함수로 정리 q,k,v를 이용해 어텐션 구하기
'''
```
